# Import Library #

```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Dataset #

```python
df = pd.read_csv("CARS_1.csv")
```

```python
df.head(10)
```

# Visualisasi Data dengan 3 GRAFIK (Pie, Bar, Line) #

# Diagram Bar/Batang #

```python
x = list(df["car_name"])
y = list(df["fuel_type"])

plt.figure(figsize=(10,10))
plt.title("Tipe bensin untuk beberapa mobil")
plt.xlabel("Nama-nama Mobil")
plt.ylabel("Tipe Bensin")
plt.bar(x, y, color="green")
plt.show()
```

## Penjelasan terhadap grafik bar diatas ##

#### koordinat y = Tipe bensin ####

#### koordinat x = Nama-nama Mobil ####

### Deskripsi ###

##### grafik bar diatas menjelaskan bahwa ada beberapa banyak nama-nama mobil di dunia dan diantaranya menggunakan tipe bensin yang bermacam-macam seperti Diesel,Petrol,Listrik dan CNG dan bisa diambil kesimpulannya bahwa tipe bensin yang paling banyak digunakan pada dataset adalah Electric/listrik dan diikuti CNG kedua, Diesel ketiga dan Petrol terakhir #####

# Diagram Pie #

```python
data_pie = df["fuel_type"].value_counts().rename_axis("fuel_type").reset_index(name="transmission_type")
```

```python
data_pie.head()
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    data_pie["transmission_type"], 
    labels=data_pie["fuel_type"],
    autopct='%.2f%%', shadow=True)
plt.legend()
plt.title("Tipe transmisi dan tipe bensinnya")
plt.show()
```

## Penjelasan terhadap grafik bar diatas ##

#### Atribut = Tipe Bensin ####

#### Label/Nilai = Tipe Tranmisi ####

### Deskripsi ###

##### grafik Pie chart diatas menjelaskan bahwa mana tipe transmisi yang memakai tipe bensin seperti Diesel,Petrol,Listrik,CNG dan bisa diambil kesimpulannya bahwa Mobil yang bertipe transmisi Automatic kebanyakan memakai tipe bensin Petrol #####

# Diagram Scatter Plot #

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='car_name',y='fuel_type',hue='transmission_type')
plt.title("Tipe Bensin untuk Berbagai macam Mobil dan Transimisnya")
plt.xlabel("Nama Mobil")
plt.ylabel("Tipe Bensin")
plt.show()
```

#### koordinat y = Tipe bensin ####

#### koordinat x = Nama Mobil ####

### Deskripsi ###

##### grafik Scatter plot diatas menunjukan bahwa Tipe Mobil mana yang memakai bensin seperti Diesel,Petrol,Listrik,CNG dan bisa diambil kesimpulannya bahwa Mobil yang bertipe transmisi Automatic kebanyakan memakai tipe bensin Petrol #####
