# Import Library

```python
import sys
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

# import dataset

```python
df = pd.read_csv('gender_classification_v7.csv')
df.head()
```

# Standarisasi Data yang akan dipakai

```python
StandardScaler().fit_transform(df[['forehead_width_cm', 'forehead_height_cm']])
```

# Clustering dengan K-Means pakai Atribut "forehead_widht_cm" dan "forehead_height_cm"

```python
model_kmeans = KMeans(n_clusters = 5)
model_kmeans.fit(df[['forehead_width_cm','forehead_height_cm']])
```

# Visualisasi Algoritma

```python
df_aktual = df[["forehead_width_cm", "forehead_height_cm", "nose_wide"]]
sns.scatterplot(x = "forehead_width_cm", y = "forehead_height_cm", data = df_aktual, hue = "nose_wide")
plt.show()
```
